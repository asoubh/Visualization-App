```python
import tkinter as tk
from tkinter.messagebox import showinfo

# module used to correcly change button bg color
import tkmacosx as MacTk

# add 
# company logo
```

```python
def analysis_type_selected(event, window):
    # # Note here that Tkinter passes an event object to onselect()
    # w = event.widget
    # index = int(w.curselection()[0])
    # value = w.get(index)
    # print('You selected item %d: "%s"' % (index, value))

    ### Parameter 1 LISTBOX ###
    # list analysis type to be chosen
    param1_lst = ('ex param 1', 'ex param 2', 'ex param 3')

    # make analysis type a tkinter variable
    param1 = tk.Variable(value=param1_lst)
    
    # create listbox
    param1_lstbx = tk.Listbox(window,
                                listvariable=param1,
                                height=6,
                                selectmode=tk.SINGLE)
    param1_lstbx.place(x=200, y=100)


##### -=-=-=- Statstical Analysis Popupwin -=-=-=- #####
def stat_analysis_popupwin():
    # assign input file path to variable
    input_file=filepath_ent.get()

    # display input spreadsheet in window - make scrollable

    # Stat Analysis Main Framework
    stat_win = tk.Toplevel(window)
    stat_win.geometry("1000x600")
    stat_win.title("Statistical Analysis Window")
    stat_win.config(bg=main_bg)

    ### Analysis Category LISTBOX ###
    # list analysis type to be chosen
    analysis_cat_lst = ('Regression', 'T-test', 'Z-test')

    # make analysis type a tkinter variable
    analysis_cat = tk.Variable(value=analysis_cat_lst)
    
    # create listbox
    analysis_cat_lstbx = tk.Listbox(stat_win,
                                    listvariable=analysis_cat,
                                    height=6,
                                    selectmode=tk.SINGLE)
    analysis_cat_lstbx.place(x=100, y=100)

    # bind click event to list box
    analysis_cat_lstbx.bind("<<ListboxSelect>>", 
                            analysis_type_selected(event=None, window=stat_win))

    ### Analyze BUTTON ###
    analyz_btn = MacTk.Button(stat_win,
                            text='Analyze!',
                            bg=main_btns,
                            activebackground='blue',
                            # analysis_cat.curselection() will only work for single selection
                            command=lambda: print([analysis_cat_lstbx.get(analysis_cat_lstbx.curselection()), 
                                                    input_file]))
    analyz_btn.place(x=100, y=200)


##### -=-=-=- Main GUI Framework -=-=-=- #####
# colors for main GUI
main_bg='#847ed4' # purple
main_btns='#c96965' # mild pink

window = tk.Tk()
window.title("Data Analysis App - DEMO")
window.geometry('500x300')
window.config(bg=main_bg)

##### Labels #####
# filepath entry
filepath_instr = tk.Label(window, 
                        text='Please copy the file path to the file you would like to analyze. \nThe file should be an excel (.xlsx) or CSV (.csv) file.', 
                        bg=main_bg, 
                        fg='black')
filepath_instr.place(x=10, y=10)
filepath_label = tk.Label(window, text='File Path:', bg=main_bg, fg='black')
filepath_label.place(x=50, y=50)

##### Entry boxes #####
# filepath entry
filepath_ent = tk.Entry(window, bg='white', fg='black')
filepath_ent.place(x=120, y=50)

##### Buttons #####
# statistical analysis button
stat_analysis_button = MacTk.Button(window,
                        text='Statistical',
                        bg=main_btns,
                        activebackground='blue',
                        command=stat_analysis_popupwin)
stat_analysis_button.place(x=100, y=170)

# visual analysis button
visual_analysis_button = MacTk.Button(window,
                        text='Graphical',
                        bg=main_btns,
                        activebackground='blue',
                        command=stat_analysis_popupwin)
visual_analysis_button.place(x=200, y=170)

# execute application in mainloop
window.mainloop()
```
